```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
list = { 'Name' : ['Ram', 'Ayush', 'Vijay','Arya', 'Arav', 'Ayush'],
        'Fruit' : ['Apple', 'Banana13', 'Grapes', None, 'Pineapple', 'Banana13'],
         'Salary' : [20000, 60000, None, 120000, 10000, 60000],
         'Age' : [18, 40, 29, None , 90, 40]
}

data = pd.DataFrame(list)
print(data)
```

```python
data.duplicated()
data.drop_duplicates(inplace=True)
print(data)
```

Are there any null values and if yess then remove them

```python
data.dtypes
```

```python
data.isnull()
data.isnull().sum()
```

```python
# for col age
data['Age'].fillna(data['Age'].mean(), inplace=True)
# for col salary
data['Salary'].replace(np.nan, 0, inplace=True)
print(data)
```

```python
data['Age'] = data['Age'].astype(int)
```

```python
print(data)
```

```python
data['Fruit'].fillna(data['Fruit'].mode(), inplace=True)
print(data)
```

```python
data.isnull().sum()
```

Normalize

```python
scaler = MinMaxScaler()
data['Salary'] = scaler.fit_transform(data[['Salary']])
print(data)
```

```python
data['Age'] = scaler.fit_transform(data[['Age']])
print(data)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
df = pd.read_csv("inconsistent_student_data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.duplicated()
df.drop_duplicates(inplace=True)
print(df)
```

```python
df.isnull()
df.isnull().sum()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')

# Impute missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Pass'] = df['Pass'].fillna(df['Pass'].mode()[0])

print(df.isnull().sum())
print(df)
```

```python
scaler = MinMaxScaler()
df[['Age', 'Marks']] = scaler.fit_transform(df[['Age', 'Marks']])
df.head()
```
